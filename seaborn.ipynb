Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.


Seaborn is a powerful python visualization library built on the top of matplotlib . It provides a high interface for drawing attractive and informative statistical graphics  .
Here are five common plots that can be created using a Seaborn along with their uses :

- Scatter Plot:
Scatter plots are useful for visualizing the relationship between two continuous variables .They help in identifying patterns , clusters , or correaltions in the data .

- Line Plot:
Line plots are effective for displaying trends over a continous interval or time series data . They are commonly used to illustrate changes in variables over time.

- Histogram:
 Histogram are ideal for visualizing the distribution of a single  continuous variable. They show the frequency or probability density of different values within the variable .
 
 
- Bar Plot :
Bar plots are suitable for comparing the values of categorical variables. They display the relationship between  categorical variable and a numerical variable by representing the latter as the height of bars .

- Heatmap :
Heatmap are useful for visualizing the magnitude of relationship between two categorical variables .They use color  intensity to represent the strength of the relationship , making them effective for identifying patterns or correlations in large datasets.

Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt


fmri_data = sns.load_dataset("fmri")
                             
                             
plt.figure(figsize=(10, 6))
sns.lineplot(data=fmri_data, x="timepoint", y="signal", hue="event", style="region", markers=True, dashes=False)
plt.title('FMRI Signal over Time for Different Events and Regions')
plt.xlabel('Timepoint')
plt.ylabel('Signal')
plt.legend(title='Event')
plt.show()
```

Q3. Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.

```python
import seaborn as sns
import matplotlib.pyplot as plt

titanic_data = sns.load_dataset("titanic")


plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
sns.boxplot(data=titanic_data, x="pclass", y="age")
plt.title("Age Distribution by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age")

plt.subplot(1, 2, 2)
sns.boxplot(data=titanic_data, x="pclass", y="fare")
plt.title("Fare Distribution by Passenger Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")

plt.tight_layout()
plt.show()
```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

```python
import seaborn as sns
import matplotlib.pyplot as plt

diamonds_data = sns.load_dataset("diamonds")


plt.figure(figsize=(10, 6))
sns.histplot(data=diamonds_data, x="price", hue="cut", kde=True, bins=30)
plt.title("Histogram of Diamond Prices by Cut")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.legend(title="Cut")
plt.show()
```
