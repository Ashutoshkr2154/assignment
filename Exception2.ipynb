Q3. What errors are defined in the ArithmeticError class? Explain any two with an example.

The ArithematicError class in python is a base class for arithematic related exceptions.It encompasses several specific error classes related to arithmetic operations .Two common errors defined in the ArithematicError class are :-
### 1. ZeroDivisionError:
- This error occurs when a division or modulo operation is attempted with zero as the divisor


```python
try :
    result = 10/0
except ZeroDivisionError as e :
    print("Error :", e)
```

### 2. OverflowError:
- This error occurs when the result of an arithematic operation exceeds the maximum representable value for a numeric type .
- Example:


```python
import sys 

try :
    result = sys.maxsize+1
except OverflowError as e :
    print("Error :", e)
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

The LookupError class in python is used as a base class for exceptions that occur when a key or index is not found during a lookup operation .It encompasses error such as KeyError and IndexError.
### 1.KeyError :

- KeyError is raised when a dictionry key is not found in the dictionary during a lookup operation .

- Example :-


```python
my_dict = {'a' :1 , 'b' :2 , 'c': 3}
try:
    value = my_dict ['d']
except KeyError as e :
    print("Error :",e)
```

### 2. IndexError :
- IndexError is raised when a sequence index (such as a list or tuple index ) is out of range or not found during a looking operation.

- Example:


```python
my_list =[1,2,3,4,5]
try:
    value = my_list[10]
except IndexError as e :
    print("Error :" , e)
```

Both KeyError and indexError are subclasses of LookupError.They indicate that a lookup operation failed due to an invalid key or index .Handling these errors approximately , such as by using try - except blocks , allows for grateful error handling and robustness in python programs.

Q5. Explain ImportError. What is ModuleNotFoundError?

'ImportError' and 'ModuleNotFoundError' are both exceptions in python that occur when there is an issue with importing a module .However , they differ in terms of their usage and the scenarios in which they are raised .

### 1. ImportError :
- 'ImportError' is a broad exception that is raised when an import statement fails for any reason .

- It occurs due to various reasons such as :-
  - The specified module or package does not exist.
  - There is an issue with the module's code , such as syntax errors or other runtime errors .
  - The module's dependencies are not met or cannot be resolved .
  
- Example :

```python
try:
    import non_existing_module
except ImportError as e :
    print("Error:" , e)
```

### 2.ModuleNotFoundError:
- 'ModuleNotFoundError' is a subclass of 'ImportError' that specially indicates that the requested module or package could not be found .
- It was introduced python 3.6 to provide more specific exception for cases where a module or package is not found .
- Example :


```python
try:
    import non_existing_module 
except ModuleNotFoundError as e :
    print("Error :" , e)
```
